```python
def post_engagement_analyzer(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 150, 300, 400]

post_engagement_analyzer(likes)
```

```python
def medicine_stock_alert(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock_alert(6)
```

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall = [70, 80, 75, 63]
required = 70

rainfall_checker(rainfall, required)
```

```python
def duplicate_account_detection(usernames):

    unique_users = set(usernames)

    if len(usernames) != len(unique_users):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

users = ["john", "alice", "john", "mike"]

duplicate_account_detection(users)
```

```python
def appointment_eligibility(age):

    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

```python
def premium_crop_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

crop_prices = [1500, 2500, 1800, 3200]

premium_crop_filter(crop_prices)
```

```python
def application_health_checker(errors):

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

application_health_checker(7)
```

```python
def transaction_limit_checker(amount):

    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_limit_checker(60000)
```

```python
def attendance_eligibility(attendance_list):

    present_count = 0

    for day in attendance_list:
        if day == 1:
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = [1, 1, 0, 1, 1]
attendance_eligibility(attendance)
```

```python
def electricity_bill_analyzer(units):

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

total_bill, usage_status = electricity_bill_analyzer(250)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
