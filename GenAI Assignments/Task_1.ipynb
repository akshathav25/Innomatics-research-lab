Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Defaulting to beginner.")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Defaulting to formal.")
        tone = "formal"

    return audience, tone
```

```python
validate_inputs("kids", "fun")
```

Task 6: Prompt Generator App

```python
templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} learners in a {tone} teaching style"
    ),

    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Create interview questions about {topic} for {audience} in a {tone} tone"
    ),

    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}
```

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    template = templates.get(style)

    if template is None:
        return "Invalid style"

    return template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt(
    "Neural Networks",
    "beginner",
    "fun",
    "storytelling"
))
```

Task 7: Template Reusability Test

```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)
```

```python
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "beginner", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Data Science", "intermediate", "casual"),
    ("Neural Networks", "expert", "formal")
]

for topic, audience, tone in inputs:
    print(reuse_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```
