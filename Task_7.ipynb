```python
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

employee = Employee(name, emp_id, department)

employee.display_id_card()
```

```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)

contact.display_contact()
```

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)

product.display_price_tag()
```

```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)

movie.display_movie()
```

```python
class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(name, address)

delivery.display_delivery()
```
