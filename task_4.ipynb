```python
def validate():
    plans=[199,299,399,599]
    while True:
        amt=int(input("enter amount:"))
        if amt>=50 and amt in plans:
            print("recharge successful")
            break
        else:
            print("invalid,retry")

validate()
```

```python
def check_stock(d):
    for k,v in d.items():
        if v<15:
            print(k,":reorder alert")
        else:
            print(k,":stock ok")

products={"pen":10,"book":25,"bag":12}
check_stock(products)
```

```python
def result(marks):
    total=0
    for m in marks:
        total+=m
    avg=total/len(marks)
    return "pass" if avg>=50 else "fail"

print(result([60,40,70]))
```

```python
def fare(km,peak):
    f=50+km*12
    if peak:
        f=f+f*0.25
    return f

while True:
    km=float(input("enter km:"))
    p=input("peak(y/n):")
    print("fare:",fare(km,p=="y"))
    if input("retry(y/n):")!="y":
        break
```

```python
def eligibility(att):
    c=0
    for a in att:
        if a=="P":
            c+=1
    per=(c/len(att))*100
    return "eligible" if per>=75 else "not eligible"

print(eligibility(["P","P","A","P"]))
```

```python
def check(p):
    d=s=0
    for ch in p:
        if ch.isdigit():
            d=1
        if ch in "@#$":
            s=1
    if len(p)>=8 and d and s:
        return "strong"
    return "weak"

print(check("pass@123"))
```
