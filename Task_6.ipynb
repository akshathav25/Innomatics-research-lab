```python
def smart_parking_system(capacity, logs):

    parked = 0
    peak_usage = 0

    # processing vehicle logs
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        # track peak usage
        if parked > peak_usage:
            peak_usage = parked

    # parking status
    if parked > capacity:
        status = "Parking Full - Alert!"
    else:
        status = "Available"

    # output
    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)
    print("Peak Parking Usage:", peak_usage)


# Sample input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```
